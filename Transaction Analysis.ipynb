### Grouping by Employee for Transactions

```python
# Creating separate df with Employee and keeping only numeric columns: Quantity, Unit Price, and Subtotal
df_trans_amt = df.drop(['Date', 'ProductID', 'ProductName', 'VendorID', 'ProductVendor', 'CustID', 'TransactionID', 'Tax(8%)', 'OrderTotal'], axis=1)
df_trans_amt
```

```python
# Whats the mean values of the numeric columns, sorted by employee?
df_mean = df_trans_amt.groupby(['Employee']).mean().round(1)
df_mean
```

***
Considering that the mean (avg) subtotal for all is $324, a majority of the employees have a higher avg sales per transaction
***

```python
# What's the sum of the numeric columns by Employee to see who's made the most sales so far?
df_sum = df_trans_amt.groupby(['Employee']).sum().round(2)
df_sum
```

```python
# Order descending values by subtotal to see who is the highest sales employee thus far
df_sum_sorted = df_sum.sort_values(by='Subtotal', ascending=False)
df_sum_sorted
```

***
These aggregations have shown me that overall, the employees have been doing a great job with their sales! 

Margaret, Nancy and Janet are the top 3 employees grossing in the highest sales.
As for Anne and Steven, the low sales could be due to recent hire or part-time.
***

### Grouping by Date (M, Q, Y) for Transactions

```python
# Extract 'Month', 'Quarter', and 'Year' and add them into columns:
df_trans_amt['Month'] = df['Date'].dt.to_period('M')  # Extract Month as YYYY-MM

df_trans_amt['Quarter'] = df['Date'].dt.to_period('Q')  # Extract Quarter as YYYY-Qx

df_trans_amt['Year'] = df['Date'].dt.year  # Extract Year
```

```python
# Grouping by Month to see Monthly Sales
total_monthly_sales = df_trans_amt.groupby(['Month'])['Subtotal'].sum().reset_index(name='Monthly Sales')
total_monthly_sales
```

*** 
Noticed that transaction data is from July 2023 to May 2025 (almost 2 years) 

From the way it's sorted by Month, can interpret a mostly steady increase in sales revenue by each month. Considering May 2025 is the lowest, it's probably only the beginning of May so I wouldn't count May as part of the analysis
***

```python
# Sorting data descending by Monthly Sales to see the highest sales month of the dataset
total_monthly_sales_sorted = total_monthly_sales.sort_values(by='Monthly Sales', ascending=False).reset_index(drop=True)
total_monthly_sales_sorted
```

***
Upon sorting the monthly sales in descending order, it can be immediately interpreted that the 1st quarter of 2025 has produced the most sales in this data. After that, it can be interpreted that Q4 of 2024 came in second place. 

What's interesting is May 2025 comes last with one of the lowest sales reported yet. Consider how well 2024 and 2025 has been going, this can be assummed that this data was pulled from early May so it's not accurate for current analysis.

To further confirm, I will group the sales by quarter
***

```python
# Grouping by Quarter to see Quarterly Sales
total_quarterly_sales = df_trans_amt.groupby(['Quarter'])['Subtotal'].sum().reset_index(name='Quarterly Sales')
total_quarterly_sales
```

```python
# Sorting quarterly sales data in descending form to see highest to lowest
total_quarterly_sales_sorted = total_quarterly_sales.sort_values(by='Quarterly Sales', ascending=False).reset_index(drop=True)
total_quarterly_sales_sorted
```

***
From sorting the sales by quarters, I can see the the first quarter is their best one yet, having a 24,000 jump from last quarter. 

I see Q2 from 2025 is appearingly not doing well according to the numbers, but since it barely started May and is only 1/3 into their quarter. Considering that information, if all looks good, Q2 of 2025 would beat Q1!

I will now group by Year to compare 2023 and 2024
***

```python
# Grouping by Year to see Yearly Sales
total_yearly_sales = df_trans_amt.groupby(['Year'])['Subtotal'].sum().reset_index(name='Yearly Sales')
total_yearly_sales
```

```python
total_yearly_sales_sorted = total_yearly_sales.sort_values(by='Yearly Sales', ascending=False).reset_index(drop=True)
total_yearly_sales_sorted
```

***
By grouping sales by the year, the business should be proud to have almost tripled their sales from 2023 to 2024.
Also considering, 2025 is barely into Q2, it's in good progress to beat 2024!

### Grouping by Employee for Transactions based on Monthly, Quarterly, and Yearly Sales

***
I will now group the sales broken into monthly, quarterly, and yearly by Employee and find the best employee for each respective timeframe
***

```python
# Grouping Monthly Sales by Employee to find best monthly employee
emp_monthly_sales = df_trans_amt.groupby(['Employee', 'Month'])['Subtotal'].sum().reset_index(name='Monthly Sales')
best_monthly_emp = emp_monthly_sales.loc[emp_monthly_sales.groupby('Month')['Monthly Sales'].idxmax()].reset_index(drop=True)
best_monthly_emp
```

***
By sorting best monthly employee, I can see Margaret has been here since the beginning and she's consistenly gave it her all. Nancy Davolio performed well month by month, coming close to beat Margaret. Overall, great team!
***

```python
# Grouping Quarterly Sales by Employee to find best quarterly employee
emp_quarterly_sales = df_trans_amt.groupby(['Employee', 'Quarter'])['Subtotal'].sum().reset_index(name='Quarterly Sales')
best_quarterly_emp = emp_quarterly_sales.loc[emp_quarterly_sales.groupby('Quarter')['Quarterly Sales'].idxmax()].reset_index(drop=True)
best_quarterly_emp
```

***
Through this, you can see that Margaret is on a roll, and she consistently took best quarterly emp in the beginning of 2024.

```python
emp_yearly_sales = df_trans_amt.groupby(['Employee', 'Year'])['Subtotal'].sum().reset_index(name='Yearly Sales')
best_yearly_emp = emp_yearly_sales.loc[emp_yearly_sales.groupby('Year')['Yearly Sales'].idxmax()].reset_index(drop=True)
best_yearly_emp
```

# Plotting Quarterly Sales from Sales and Employee View

```python
# Group by Quarter and Employee, then sum the sales
quarterly_sales = df_trans_amt.groupby(['Quarter', 'Employee'])['Subtotal'].sum().unstack().fillna(0)

# Plotting
quarterly_sales.plot(kind='bar', figsize=(12, 8), stacked=True, colormap='tab20')

# Adding title and labels
plt.title('Quarterly Sales Review by Employee (Stacked)', fontsize=14)
plt.xlabel('Quarters', fontsize=12)
plt.ylabel('Total Sales', fontsize=12)
plt.xticks(rotation=30)  # Rotate the quarter labels for better readability
plt.legend(title='Employee', bbox_to_anchor=(1, 1), loc='upper left')  # Move legend outside the plot

# Show the plot
plt.show()
```

***
I chose to go with a stacked bar chart, because I feel like its a great way to show visually the growth of sales by quarterly from July 2023 to May 2025. You can clearly see the upward trend by quarter.

While doing so, you can also clearly distinguish how each employee contributed to each quarterly sales by seeing their assigned color in each bar.

I feel like this was a tasteful way to show the data, rather than have two bar charts (one for quarterly sales and other for best employee by quarterly sales)

Based on our analysis, we can interpret that the business is doing really well and had a huge jump in their sales going into 2025. 

We can also interpret that Margaret Peacock is their best employee who brings in consistent great sales. As shown above, Nancy and Janet, are trailing close to Margaret so if this pattern is projected to continue, 2025 would be the company's best year by far.
